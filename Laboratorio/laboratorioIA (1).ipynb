**NORMALIZACION DE CARACTERISTICAS**

```python
#NORMALIZACION DE CARACTERISTICAS
def  featureNormalize(X):

    X_norm = X.copy()
    mu = np.zeros(X.shape[1])
    sigma = np.zeros(X.shape[1])

    mu = np.mean(X, axis = 0)
    sigma = np.std(X, axis = 0)
    X_norm = (X - mu) / sigma

    return X_norm, mu, sigma
```

```python
#CALCULAR COSTO
def computeCostMulti(X, y, theta):
    # Inicializa algunos valores utiles
    m = y.shape[0] # numero de ejemplos de entrenamiento

    J = 0

    # h = np.dot(X, theta)

    J = (1/(2 * m)) * np.sum(np.square(np.dot(X, theta) - y))

    return J
```

```python
theta = np.zeros((X_norm.shape[1], 1))

# Suponiendo que y es un DataFrame y quieres obtener la primera columna como array
y_array = y.iloc[:, 0].values.reshape(-1, 1)


# Ahora puedes usar y_array en lugar de y
JJ = computeCostMulti(X_norm, y_array, theta)


# Imprimir el resultado
print(f"Con theta: {theta} se obtiene un costo inicial de: {JJ}")
```

**DESCENSO GRADIENTE**

```python
#DESCENSO GRADIENTE
def gradientDescentMulti(X, y, theta, alpha, num_iters):

    # Inicializa algunos valores
    m = y.shape[0] # numero de ejemplos de entrenamiento

    # realiza una copia de theta, el cual será acutalizada por el descenso por el gradiente

    theta = theta.copy()

    J_history = []

    for i in range(num_iters):
        theta = theta - (alpha / m) * np.dot(X.T, (np.dot(X, theta) - y))

        J_history.append(computeCostMulti(X, y, theta))

    return theta, J_history
```

```python
# Elegir algun valor para alpha (probar varias alternativas)
alpha = 0.001 # alpha = 0.003
num_iters = 11000

#ejecutar el descenso por el gradiente
theta, J_history = gradientDescentMulti(X_norm, y_array, theta, alpha, num_iters)
# Muestra los resultados del descenso por el gradiente
print(f'theta calculado por el descenso por el gradiente: {theta}')

# Grafica la convergencia del costo (historial de la funcion costo)
pyplot.plot(np.arange(len(J_history)), J_history, lw=2)
pyplot.xlabel('Numero de iteraciones')
pyplot.ylabel('Costo J')
pyplot.title('Convergencia del costo')
pyplot.show()




```
